## Las variables
1. NUM --> Valor para determinar su signo
2. R --> Resultado del signo del valor
## Formulas
1. $R=$  $NUM\ge0$

```python
# Para cuando es verdadero (si)
# Calculando numeros positivos negativos y positivos con if else
#  Datos de entrada NUM
# Mostrar numero negativo si es falso
# Si es verdadero mostrar positivo

NUM= float(input("Ingrese el numero: "))
if NUM >= 0:
    R= "positivo"
else:
    R= "negativo"
print("El numero ingresado es: ", R)
print("Fin del programa")
print("Programa realizado por Iram Jimenez NL = 28")
```

```python
# Para cuando es falso (no)
# Calculando numeros positivos negativos y positivos con if else
#  Datos de entrada NUM
# Mostrar numero negativo si es falso
# Si es verdadero mostrar positivo

NUM= float(input("Ingrese el numero: "))
if NUM >= 0:
    R= "positivo"
else:
    R= "negativo"
print("El numero ingresado es: ", R)
print("Fin del programa")
print("Programa realizado por Iram Jimenez NL = 28")
```
